```python
import pandas as pd
from scipy.stats import shapiro
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

model_names = ["meta-llama_Llama-3.2-3B", "facebook_xglm-4.5B", "deepseek-ai_deepseek-llm-7b-base", "TsinghuaAI_CPM-Generate"]
```

```python
filename = f'result/2_{model_names[0]}_results_apx.csv'
df = pd.read_csv(filename, on_bad_lines='warn')

# Calculate the average APX
df['apx_avg'] = df[['apx_1', 'apx_2', 'apx_3']].mean(axis=1)

# Clean data: remove NaNs and inf values
data = df['apx_avg'].replace([float('inf'), -float('inf')], pd.NA).dropna()

# === Shapiro-Wilk Test for Average APX ===
print(f"\n=== Shapiro-Wilk Test for apx_avg ===")
stat, p = shapiro(data)
print(f"Statistic = {stat:.4f}, p-value = {p:.4e}")
if p > 0.05:
    print("✅ Data appears normally distributed.")
else:
    print("❌ Data does not appear normally distributed.")

# Histogram with KDE
plt.figure(figsize=(6, 4))
sns.histplot(data, kde=True, bins=100)
plt.title("Histogram of Average APX - llama")
plt.xlabel("apx_avg")
plt.ylabel("Frequency")
plt.xlim(0, 10000)
plt.tight_layout()
plt.show()

# Q-Q Plot
plt.figure(figsize=(6, 4))
stats.probplot(data, dist="norm", plot=plt)
plt.title("Q-Q Plot of Average APX")
plt.tight_layout()
plt.show()
```
